```python
#import random
from itertools import permutations
import copy
from collections import Counter
import pandas as pd
import time

# beans = [["衣索匹亞", 16, 400, 3], 
#          ["肯亞", 18, 450, 1], 
#          ["馬拉威", 17, 350, 3], 
#          ["瓜地馬拉", 17, 380, 2], 
#          ["哥斯大黎加", 18, 390, 2], 
#          ["尼加拉瓜", 16, 350, 3], 
#          ["宏都拉斯", 17, 380, 2], 
#          ["巴拿馬", 18, 450, 2], 
#          ["巴西", 16, 350, 3], 
#          ["哥倫比亞", 17, 380, 2], 
#          ["曼特寧", 20, 330, 4]]

beans = [["衣索匹亞", 16, 400, 3], 
         ["肯亞", 18, 450, 1], 
         ["瓜地馬拉", 17, 380, 2], 
         ["哥斯大黎加", 18, 390, 2], 
         ["巴拿馬", 18, 450, 2],
         ["巴西", 16, 350, 3],
         ["哥倫比亞", 17, 380, 2], 
         ["曼特寧", 20, 330, 4]]
```

```python
# BFS: Breadth-First Search
start_time = time.time()

print("=============BFS=================")
roasted_df = []
time_df =[]
price_df =[]
j_df = []
bean_item_df = []

for beans_set in list(permutations(beans)):
    working_time = 100
    roast_time = 0
    roast_time_last = 0
    price = 0
    price_last = 0
    roasted = []
    j = 0
    #print(beans_set)
    beans_set_roast = copy.deepcopy(beans_set)
    #print(beans_set_roast)
    while working_time > roast_time:
        for i in beans_set_roast:
            if roast_time > working_time: 
                break
            #print(i)
            if i[3] > 0:
                roast_time_last = i[1]
                roast_time = roast_time + i[1]
                price_last = i[2]
                price = price + i[2]
                roasted.append(i[0])
                i[3] = i[3] - 1
                j = j + 1        
    roasted = roasted[:-1]
    roasted_sum = Counter(roasted)
    roast_time = roast_time - roast_time_last
    price = price - price_last
    j = j - 1
    roasted_df.append(roasted_sum)
    bean_item_df.append(set(roasted))
    time_df.append(roast_time)
    price_df.append(price)
    j_df.append(j)
    df = pd.DataFrame({'烘焙項目':roasted_df,
                       '豆子種類':bean_item_df,
                       '總鍋數': j_df, 
                       '總烘焙時間': time_df ,
                       '價格':price_df})

print("--- %s seconds ---" % (time.time() - start_time))    
df.sort_values("價格")
df.to_csv("BFS.csv", sep=',', encoding='big5')
#df[['烘焙項目', '豆子種類', '總鍋數', '總烘焙時間', '價格']]
```

```python
start_time = time.time()

# DFS: Depth-First Search
roasted_df = []
time_df =[]
price_df =[]
j_df = []
bean_item_df = []

print("=============DFS=================")
for beans_set in list(permutations(beans)):
    working_time = 100
    roast_time = 0
    roast_time_last = 0
    price = 0
    price_last = 0
    roasted = []
    j = 0
    #print(beans_set)
    beans_set_roast = copy.deepcopy(beans_set)
    #print(beans_set_roast)
    while working_time > roast_time:
        for i in beans_set_roast:
            #if roast_time > working_time: 
            #    break
            #print(i)
            while i[3] > 0:
                if roast_time > working_time: break
                roast_time_last = i[1]
                roast_time = roast_time + i[1]
                price_last = i[2]
                price = price + i[2]
                roasted.append(i[0])
                i[3] = i[3] - 1
                j = j + 1        
    roasted = roasted[:-1]
    roasted_sum = Counter(roasted)
    roast_time = roast_time - roast_time_last
    price = price - price_last
    j = j - 1
    roasted_df.append(roasted_sum)
    bean_item_df.append(set(roasted))
    time_df.append(roast_time)
    price_df.append(price)
    j_df.append(j)
    df = pd.DataFrame({'烘焙項目':roasted_df,
                       '豆子種類':bean_item_df,
                       '總鍋數': j_df, 
                       '總烘焙時間': time_df ,
                       '價格':price_df})

print("--- %s seconds ---" % (time.time() - start_time))
df.sort_values("價格")
df.to_csv("DFS.csv", sep=',', encoding = 'big5')
#df[['烘焙項目', '豆子種類', '總鍋數', '總烘焙時間', '價格']]
```
